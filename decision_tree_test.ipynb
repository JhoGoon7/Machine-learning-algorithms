```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
dataset = load_iris()
```

```python
dataset
```

```python
data = pd.DataFrame(data = dataset.data , columns = dataset.feature_names)
```

```python
data.head()
```

```python
data["Species"] = dataset.target
```

```python
data.head()
```

```python
data["Species"].value_counts()
```

```python
X = data.iloc[:,0:4]
```

```python
X.head()
```

```python
y = data.iloc[:,4]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.get_params()
```

```python
dtc.fit(X_train,y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
y_test
```

```python
y_pred
```

```python
data.info()
```

```python
#the gini is the purity factor , the more the gini is big the more the sample is impure , use it with a tree that is composed by a dominant class.
#the entropy is used if we want to maximize information gain, more sensitive to class imbalance
```

```python
from sklearn.tree import plot_tree

feature_names = dataset.feature_names 
class_names = dataset.target_names

plt.figure(figsize=(20, 10))
plot_tree(dtc, feature_names=feature_names, class_names=class_names, filled=True, rounded=True)
plt.show()
```
