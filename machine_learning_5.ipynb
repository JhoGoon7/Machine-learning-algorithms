```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.datasets import load_breast_cancer 
```

```python
data = load_breast_cancer()
```

```python
dataset = pd.DataFrame(data.data, columns = data.feature_names)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset['target'] = data.target
```

```python
X = dataset.drop(columns = ['target'])
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
#let's normalise the data : it means that we the values of some features have to be modify to be compared with the others ,because all te features have
#not the same unity  
```

```python
#create a scaler object
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
y = dataset.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X_scaled , y, test_size = 0.3 , random_state = 42)
```

```python
#already convert X_train and executed with succes
#X_test = X_test.to_numpy()
#y_train = y_train.to_numpy()
#y_test = y_test.to_numpy()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()

#train the model on the training data
lr.fit(X_train , y_train)
```

```python
#predict the target varibale based on the test data
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
#evaluation of the model
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test , y_pred)
print(f"Accuracy : {accuracy : .2f}")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test , y_pred))
```
