```python
#above we have found only one centroid , we got the coordinate for only one centroid so one cluster but we need multiple cluster so more centroids
def random_centroid(dataset , k):
    centroids = []
    for i in range(k):
        centroid  = dataset0.apply(lambda x: float(x.sample().to_numpy()[0]))
        centroids.append(centroid)
    return pd.concat(centroids , axis=1) #to concatenate the different pandas series in columns
```

```python
centroids = random_centroid(dataset0 , 5)
centroids
```

```python
#Label each data point
np.sqrt(((dataset0 - centroids.iloc[:,0]) ** 2).sum(axis=1))  #the iloc fucntion is used to acces row and columns from a DataFrame or a series
```

```python
def get_labels(dataset0 , centroids):
    distance = centroids.apply(lambda x: np.sqrt(((dataset0 - x) ** 2).sum(axis=1)))
    return distance.idxmin(axis=1)    #it gives directlythe the index where the value is the smallest
#we see that the player 99 is nearest to the cluster 1 , because the distance between those them is the smallest
```

```python
label = get_labels(dataset0 , centroids)
dataset0.head()
```

```python
#updating the centroid
def new_centroid(dataset0 , label , k):
    return dataset0.groupby(label).apply(lambda x: np.exp(np.log(x).mean())).T
```

```python
#plot the updating of each centroids during the process
from sklearn.decomposition import PCA #help to visulize the data , because we re in 5 dimension but it will arange the dimension
import matplotlib.pyplot as plt
from IPython.display import clear_output #it will help by clearing the notebook and print the new graph each iterations
```

```python
def plot_clusters(dataset0 , label , centroids , iteration):
    pca = PCA(n_components = 2) #to tell to return our data of 5 columns into 2 columns 
    data_2d = pca.fit_transform(dataset0) #transform the data and then execute
    centroid_2d = pca.transform(centroid.T)
    clear_output(wait=True)
    plt.title(f'Iteration number {iteration}')
    plt.scatter(x=data_2d[:,0] , y=data_2d[:,1] , c=label)
    plt.scatter(x=centroid_2d[:,0] , y=centroid_2d[:,1])
    plt.show()
```

```python
max_iterations = 100
k = 3

centroid = random_centroid(dataset0 , k)
old_centroid = pd.DataFrame()
iteration = 3

while iteration < max_iterations and not centroid.equals(old_centroid):
    old_centroid = centroid
    label = get_labels(dataset0 , centroid)
    centroid = new_centroid(dataset0 , label , k)
    plot_clusters(dataset0 , label , centroid , iteration)
    iteration += 1
    
    
```
